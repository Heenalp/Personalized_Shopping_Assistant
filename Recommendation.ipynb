```python
from flask import Flask, request, jsonify, render_template
import os
import pandas as pd
import numpy as np
import re
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.chat.util import Chat, reflections

# Download NLTK stopwords and WordNet (if not already downloaded)
nltk.download('stopwords')
nltk.download('wordnet')

# Set environment variable for threading
os.environ["OMP_NUM_THREADS"] = "2"

# Print the current working directory
print("Current Working Directory:", os.getcwd())

# Load the dataset from an Excel file
file_path = r"C:\Users\heena\Desktop\Recommendation Project\online_retail.xlsx"  # Update this path

# Check if the file exists
if not os.path.exists(file_path):
    raise FileNotFoundError(f"The file '{file_path}' does not exist. Please check the file path.")

new_df_cleaned = pd.read_excel(file_path, engine="openpyxl")

# Drop rows with missing values in key columns
new_df_cleaned = new_df_cleaned.dropna(subset=["Customer ID", "Quantity", "Price"])
new_df_cleaned["Customer ID"] = new_df_cleaned["Customer ID"].astype(int)

# Aggregate total spending per customer
customer_data = new_df_cleaned.groupby("Customer ID").agg({'Quantity': 'sum', "Price": 'sum'})
customer_data['Total_Spending'] = customer_data["Quantity"] * customer_data["Price"]

# Handle outliers (optional - removes top 1% high spenders)
customer_data = customer_data[customer_data['Total_Spending'] < customer_data['Total_Spending'].quantile(0.99)]

# Ensure Total_Spending is not None or NaN
customer_data['Total_Spending'].fillna(0, inplace=True)

# Standardize the data
scaler = StandardScaler()
customer_data['Total_Spending_Scaled'] = scaler.fit_transform(customer_data[['Total_Spending']])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(customer_data[['Total_Spending_Scaled']])

# Create a pivot table: customers vs products (interaction matrix)
customer_product_matrix = new_df_cleaned.pivot_table(index="Customer ID", 
                                                     columns="Description", 
                                                     values="Quantity", 
                                                     fill_value=0)

# Load the reference descriptions from Ids.xlsx
reference_descriptions = pd.read_excel(r"C:\Users\heena\Desktop\Recommendation Project\Ids.xlsx", engine="openpyxl")["Description"].dropna().unique()

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

# Function to clean and preprocess descriptions
def clean_description(description):
    description = description.lower()
    description = re.sub(r"[^a-zA-Z\s]", "", description)
    description = re.sub(r"\s+", " ", description).strip()
    description = " ".join([lemmatizer.lemmatize(word) for word in description.split() if word not in stop_words])
    return description

# Apply cleaning to the reference descriptions
cleaned_reference_descriptions = [clean_description(desc) for desc in reference_descriptions]

# Create a mapping from cleaned descriptions to original descriptions
description_mapping = {clean_desc: orig_desc for clean_desc, orig_desc in zip(cleaned_reference_descriptions, reference_descriptions)}

# Clean the descriptions in new_df_cleaned
new_df_cleaned["Cleaned_Description"] = new_df_cleaned["Description"].apply(clean_description)

# Map cleaned descriptions back to original descriptions
new_df_cleaned["Original_Description"] = new_df_cleaned["Cleaned_Description"].map(description_mapping)

# Fill missing original descriptions with cleaned descriptions (if no mapping exists)
new_df_cleaned["Original_Description"].fillna(new_df_cleaned["Cleaned_Description"], inplace=True)

# Extract cleaned descriptions for analysis
cleaned_descriptions = new_df_cleaned["Cleaned_Description"].unique()

# Create a TF-IDF vectorizer
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(cleaned_descriptions)

# Compute cosine similarity between products
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create a mapping from cleaned description to index
desc_to_index = {desc: idx for idx, desc in enumerate(cleaned_descriptions)}
index_to_desc = {idx: desc for idx, desc in enumerate(cleaned_descriptions)}

# Extract all unique product descriptions from the dataset and clean them
all_descriptions = new_df_cleaned["Description"].apply(clean_description).unique()

# Get the most popular products (most sold products)
most_sold_products = new_df_cleaned.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(5).index.tolist()

# Define chatbot responses for product trends and recommendations
pairs = [
    # Gift recommendations
    (r"(.*)gift ideas(.*)|(.*)recommend for gifts(.*)", [
        f"Great gift ideas include {np.random.choice(all_descriptions, size=2, replace=False)}.",
        f"Personalized gifts like {np.random.choice(all_descriptions, size=2, replace=False)} are always a hit!"
    ]),
    
    # Popular products
    (r"(.*)popular(.*)|(.*)most popular(.*)", [
        f"Our most popular items are: {', '.join(most_sold_products)}.",
        f"Check out our bestsellers like: {', '.join(most_sold_products)}."
    ]),
    
    # Personalized recommendations
    (r"(.*)recommend for me(.*)|(.*)personalized(.*)", [
        "Sure! Please enter your Customer ID to get personalized recommendations."
    ]),
    
    # Thank you
    (r"(.*)thank you(.*)", [
        "You're welcome! Let me know if you need more recommendations.",
        "Glad to help! Feel free to ask more questions."
    ]),
    
    # Greetings
    (r"hello|hi|hey", [
        "Hello! How can I assist you with product recommendations today?",
        "Hi there! What products or trends are you interested in?"
    ]),
    (r"hello(.*)|hi(.*)|hey(.*)", [
        "Hello! Let me help you find the perfect products.",
        "Hi! What can I help you with today?"
    ]),
    
    # Exit
    (r"(.*)exit(.*)", [
        "Goodbye! Happy shopping and visit us again soon!",
        "See you later! Don't forget to check out our latest products."
    ]),
    
    # Fallback response for unknown queries
    (r".*", [
        "I'm sorry, I didn't understand that. Here are some things I can help with:",
        "1. What are good gift ideas?",
        "2. What are your most popular products?",
        "3. Can you recommend products for me? (Personalized recommendations)",
        "Please ask one of the above questions, or type 'exit' to end the chat."
    ]),
]

# Create chatbot instance
product_bot = Chat(pairs, reflections)

# Function to recommend products based on collaborative filtering
def recommend_products_collaborative(customer_id, num_recommendations=10):  # Set default to 10
    if customer_id not in customer_data.index:
        raise ValueError(f"Customer ID {customer_id} not found!")
    
    target_cluster = customer_data.loc[customer_id, "Cluster"]
    same_cluster_customers = customer_data[customer_data["Cluster"] == target_cluster].index
    filtered_customer_product_matrix = customer_product_matrix.loc[same_cluster_customers]

    similarity_matrix = cosine_similarity(filtered_customer_product_matrix)

    if customer_id not in filtered_customer_product_matrix.index:
        raise ValueError(f"Customer ID {customer_id} not found in cluster data!")

    customer_idx = np.where(filtered_customer_product_matrix.index == customer_id)[0][0]
    similar_customers = np.argsort(similarity_matrix[customer_idx])[::-1][1:num_recommendations+1]
    
    recommended_products = []
    for similar_customer in similar_customers:
        similar_customer_id = filtered_customer_product_matrix.index[similar_customer]
        purchased_products = new_df_cleaned[new_df_cleaned["Customer ID"] == similar_customer_id]["Description"].unique()
        recommended_products.extend(purchased_products)

    # Ensure only 10 unique recommendations are returned
    return list(set(recommended_products))[:num_recommendations]

# Function to recommend products based on content-based filtering
def recommend_products_content_based(customer_id, num_recommendations=10):  # Set default to 10
    purchased_products = new_df_cleaned[new_df_cleaned["Customer ID"] == customer_id]["Cleaned_Description"].unique()
    
    if len(purchased_products) == 0:
        return []

    purchased_indices = [desc_to_index[product] for product in purchased_products if product in desc_to_index]
    avg_similarity = cosine_sim[purchased_indices].mean(axis=0)
    similar_products = np.argsort(avg_similarity)[::-1]
    similar_products = [idx for idx in similar_products if idx not in purchased_indices]

    recommended_products_cleaned = [index_to_desc[idx] for idx in similar_products[:num_recommendations]]
    recommended_products_original = [description_mapping[product] for product in recommended_products_cleaned]

    return recommended_products_original[:num_recommendations]  # Ensure only 10 recommendations

# Initialize Flask app
app = Flask(__name__, template_folder="Design")

# Root route - Render the HTML form
@app.route('/')
def home():
    return render_template('index.html')

# Recommendation endpoint - Handle form submission
@app.route('/recommend', methods=['GET'])
def recommend():
    try:
        customer_id = request.args.get('customer_id')  # Get the customer_id from the form
        print(f"Customer ID: {customer_id}")  # Debugging
        
        if not customer_id:
            return render_template('index.html', error="Please provide a Customer ID.")
        
        # Validate if customer_id is a valid integer
        if not customer_id.isdigit():
            return render_template('index.html', error="Invalid Customer ID. Please enter a valid number.")
        
        customer_id = int(customer_id)  # Convert to integer
        print(f"Customer ID (int): {customer_id}")  # Debugging
        
        # Get 10 collaborative recommendations
        collaborative_recommendations = recommend_products_collaborative(customer_id, num_recommendations=10)
        print(f"Collaborative Recommendations: {collaborative_recommendations}")  # Debugging
        
        # Get 10 content-based recommendations
        content_based_recommendations = recommend_products_content_based(customer_id, num_recommendations=10)
        print(f"Content-Based Recommendations: {content_based_recommendations}")  # Debugging
        
        # Pass the recommendations to the template
        return render_template('index.html', 
                              customer_id=customer_id,
                              recommendations={
                                  "collaborative": collaborative_recommendations,
                                  "content_based": content_based_recommendations
                              })
    except ValueError as e:
        print(f"Error: {e}")  # Debugging
        return render_template('index.html', error="Invalid Customer ID. Please enter a valid number.")
    except Exception as e:
        print(f"Error: {e}")  # Debugging
        return render_template('index.html', error=str(e))

# Trend advice endpoint
@app.route('/trend', methods=['GET'])
def trend():
    try:
        user_input = request.args.get('query')
        response = product_bot.respond(user_input)
        return jsonify({"Trend Advice": response})
    except Exception as e:
        return jsonify({"error": str(e)}), 400

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=False)  # Disable debug mode to avoid SystemExit error
```
